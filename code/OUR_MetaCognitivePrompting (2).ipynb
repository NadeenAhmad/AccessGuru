```python
from openai import OpenAI
import os

# Initialize the OpenAI client with your API key
#client = OpenAI(api_key='API-KEY')

# Define a function to manage an accessibility issues chat
def fix_accessibility_issues_mp_chat():
    # Start with an empty message history
    chat_history = [
        {"role": "system", "content": "You are a web accessibility expert."}
    ]

    # Function to add a prompt and get a response
    def send_prompt(prompt):
        # Add user's prompt to the history
        chat_history.append({"role": "user", "content": prompt})

        # Get a response from the model
        response = client.chat.completions.create(
            model="gpt-4-0125-preview",
            messages=chat_history
        )

        # Append the model's response to the history
        # Access the 'content' attribute directly from the ChatCompletionMessage object
        chat_response = response.choices[0].message.content
        chat_history.append({"role": "assistant", "content": chat_response})

        # Return the model's response
        return chat_response

    # Function to reset the chat history
    def reset():
        nonlocal chat_history
        chat_history = [
            {"role": "system", "content": "You are a web accessibility expert."}
        ]

    # Return the functions to handle the chat and reset the session
    return send_prompt, reset

# Create a chat session
#chat_session, reset_session = fix_accessibility_issues_mp_chat()

# Example usage:
# Send a prompt
#response = chat_session("How can I make my website more accessible?")
#print(response)

# Reset the chat session
#reset_session()

# After reset, send another prompt
#response = chat_session("What are the best practices for web accessibility?")
#print(response)
```

```python
def fix_accessibility_issues_mp_chat_full(URL, violationType, violationDescription, impact, HTMLElement, guidelines):
  # Create a chat session
  chat_session, reset_session = fix_accessibility_issues_mp_chat()
  prompt_comprehension_clarification= """
  Clarify your understanding of the web accessibility violation: Web page URL:  "$URL$", Violation type: "$violationType$",
  Violation description: "$violationDescription$", Impact: "$impact$", Affected HTML Element(s): "$HTMLElement$"
  """
  prompt_comprehension_clarification_full = prompt_comprehension_clarification.replace("$URL$", URL).replace("$violationType$", violationType).replace("$violationDescription$", violationDescription).replace("$impact$", impact).replace("$HTMLElement$", HTMLElement)

  prompt_prelimnary_judgment = """Based on your understanding, provide a preliminary correction for the web accessibility violation based on the following WCAG guidelines "$guidelines$". """
  prompt_prelimnary_judgment_full = prompt_prelimnary_judgment.replace("$guidelines$", guidelines)

  #prompt_critical_evaluation = "Critically assess your preliminary correction. If you are unsure about the initial correction, try to reassess it."
  #mitigation
  prompt_critical_evaluation = "Critically assess your preliminary correction, make sure to correct the initial web accessibility violation without introducing new web accessibility violations. If you are unsure about the initial correction, try to reassess it."
  prompt_decision_confirmation = """Confirm your final decision on whether the correction is accurate or not and provide the reasoning for your decision.
  Enclose the corrected HTML code only to replace the initial code with violations between these two marker strings: \"###albidaya###\" as first line and \"###alnihaya###\" as last line.

  """

  prompt_confidence_level= """
  Evaluate your confidence (0-100%) in your correction, enclose your confidence score between these two marker strings: \"###albidaya###\" as first line and \"###alnihaya###\" as last line.
  Provide an explanation for this confidence level, enclose your explanation between these two marker strings: \"###albidaya2###\" as first line and \"###alnihaya2###\" as last line.
  """

  #Chat session
  print("---------------------------------Comprehension Clarification---------------------------------")
  print(chat_session(prompt_comprehension_clarification_full))
  chat_session(prompt_comprehension_clarification_full)
  print("---------------------------------Preliminary Judgement---------------------------------")
  print(chat_session(prompt_prelimnary_judgment_full))
  chat_session(prompt_prelimnary_judgment_full)
  print("---------------------------------Critical Evaluation---------------------------------")
  print(chat_session(prompt_critical_evaluation))
  chat_session(prompt_critical_evaluation)
  print("---------------------------------Decision Confirmation---------------------------------")
  final_decision = chat_session(prompt_decision_confirmation)
  start = final_decision.index("###albidaya###")+ len("###albidaya###")
  end = final_decision.index("###alnihaya###", start)
  print(final_decision[start:end])
  print(final_decision)
  print("---------------------------------Confidence Level---------------------------------")
  print(chat_session(prompt_confidence_level))
  confidence_reason = chat_session(prompt_confidence_level)
  start21 = confidence_reason.index("###albidaya###")+ len("###albidaya###")
  end21 = confidence_reason.index("###alnihaya###", start21)
  print(confidence_reason[start21:end21])

  start2 = confidence_reason.index("###albidaya2###")+ len("###albidaya2###")
  end2 = confidence_reason.index("###alnihaya2###", start2)
  print(confidence_reason[start2:end2])



  reset_session()

  return final_decision[start:end], confidence_reason[start21:end21], confidence_reason[start2:end2]
```

```python
def get_guidelines(violationType):
  guide = []
  json_file_path = '/content/WebAccessibility/notebooks/mapping_dict_file.json'
  # Open and read the JSON file
  with open(json_file_path, 'r') as file:
     data = json.load(file)
  guidelineNames=data[violationType]
  guidelines= pd.read_csv('/content/WebAccessibility/data/TrustedGuidelines.csv')
  for x in guidelineNames:
      guide.append(str(x+":" +guidelines[guidelines['guideId'] == x]['description'].values[0]))
  return guide


```
